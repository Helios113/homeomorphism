# Sanity 4: Intrinsic Dimension on Known Manifold
Generates a noisy 1D circle manifold in high dimension and checks TwoNN/ESS behavior.

```python
import torch
import matplotlib.pyplot as plt
from homeomorphism.id_est import estimate_id

torch.manual_seed(0)
N, d = 1000, 64
theta = 2 * torch.pi * torch.rand(N)
X = torch.zeros(N, d)
X[:, 0] = torch.cos(theta)
X[:, 1] = torch.sin(theta)
X[:, 2:] = 0.03 * torch.randn(N, d - 2)

id_twonn = estimate_id(X, 'twonn')
id_ess = estimate_id(X, 'ess')
print(f'TwoNN ID: {id_twonn:.3f}')
print(f'ESS ID  : {id_ess:.3f}')

plt.figure(figsize=(5, 5))
plt.scatter(X[:, 0].numpy(), X[:, 1].numpy(), s=6, alpha=0.5)
plt.axis('equal')
plt.title('Projected synthetic manifold (first 2 dims)')
plt.tight_layout()
plt.show()
```
